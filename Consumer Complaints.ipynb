```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Consumer_Complaints.csv')
```

```python
data
```

```python
data = data[['Product', 'Consumer Complaint']]
```

```python
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data
```

```python
prod_categ = list(data.Product.unique())
```

```python
len(prod_categ)
```

```python
prod_category = []
for prod in data['Product']:
    prod_category.append(prod_categ.index(prod))
```

```python
data['prod_category'] = prod_category
```

```python
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(sublinear_tf=True,  # to scale the data
                        norm='l2',  # l1 and l2 regurlarization
                        min_df=5,  # words with lower frequency counts than the threshold(5)
                        stop_words='english',  # remove stop words
                        ngram_range=(1,2))  # to include unigrams(only one word) and bigrams(consists of two words)
```

```python
features = tfidf.fit_transform(data['Consumer Complaint'][:10000])
labels = data.prod_category[:10000]
```

```python
features = features.toarray()
```

```python
features.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
model = LinearSVC()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

### Try out some complaints on this model

```python
complaint = [input('Enter compliant: ')]  # o/p: Credit Card
my_predict = model.predict(tfidf.transform(complaint).toarray())
print('\n\nProduct: ',prod_categ[int(my_predict)])
```
